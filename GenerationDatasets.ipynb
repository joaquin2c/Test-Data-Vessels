```python
def f_list_patient(list_imgs):
    list_patient = []
    temp = ''
    for name in list_imgs:
        patient = name.split(".")[0].split("_")[0]
        if patient != temp :
            list_patient.append(patient)
        temp = patient
    return list_patient

def f_list_3cut(slide):
    list_slide = []
    n_2 = slide.split(".")[0].split("_")[1].split("-")[1]
    n_1 = str(int(n_2) - 1).zfill(3)
    n_3 = str(int(n_2) + 1).zfill(3)
    list_slide.append(slide.split(".")[0].split("_")[0] + '_1-' + n_1 + '.npy')
    list_slide.append(slide.split(".")[0].split("_")[0] + '_1-' + n_3 + '.npy')
    return list_slide
```

```python
def box_slide(slide):
    dim = slide.split(";")

    if int(dim[5]) <= 256 and int(dim[6]) <= 256 :
        d_x = np.round((256 - int(dim[5]))/2)
        d_y = np.round((256 - int(dim[6]))/2)
        if int(dim[1]) > d_x :
            x_min = int(dim[1]) - d_x
            x_max = x_min + 255
            if x_max > 511:
                x_max = 511
                x_min = 511 - 255
        else:
            x_min = int(dim[1])
            x_max = x_min + 255
            if x_max > 511:
                x_max = 511
                x_min = 511 - 255
        if int(dim[3]) > d_y :
            y_min = int(dim[3]) - d_y
            y_max = y_min + 255
            if y_max > 511:
                y_max = 511
                y_min = 511 - 255
        else:
            y_min = int(dim[3])
            y_max = y_min + 255
            if y_max > 511:
                y_max = 511
                y_min = 511 - 255
    else:
        if int(dim[5]) > int(dim[6]):
            d_x = 25 # Cuanto mas?
            d_y = np.round(0.5*(int(dim[6])-int(dim[5]))) + d_x
            if int(dim[1]) > d_x :
                x_min = int(dim[1]) - d_x
                x_max = x_min + int(dim[5]) + 2*d_x
                if x_max > 511:
                    x_max = 511
                    x_min = 511 - int(dim[5]) - 2*d_x
            else:
                x_min = int(dim[1])
                x_max = x_min + int(dim[5]) + 2*d_x
                if x_max > 511:
                    x_max = 511
                    x_min = 511 - int(dim[5]) - 2*d_x
            if int(dim[3]) > d_y :
                y_min = int(dim[3]) - d_y
                y_max = y_min + int(dim[5]) + 2*d_x
                if y_max > 511:
                    y_max = 511
                    y_min = 511 - int(dim[5]) - 2*d_x
            else:
                y_min = int(dim[3])
                y_max = y_min + int(dim[5]) + 2*d_x
                if y_max > 511:
                    y_max = 511
                    y_min = 511 - int(dim[5]) - 2*d_x
        else:
            d_y = 25 # Cuanto mas?
            d_x = np.round(0.5*(int(dim[5])-int(dim[6]))) + d_y
            if int(dim[1]) > d_x :
                x_min = int(dim[1]) - d_x
                x_max = x_min + int(dim[6]) + 2*d_y
                if x_max > 511:
                    x_max = 511
                    x_min = 511 - int(dim[6]) - 2*d_y
            else:
                x_min = dim[1]
                x_max = x_min + int(dim[6]) + 2*d_y
                if x_max > 511:
                    x_max = 511
                    x_min = 511 - int(dim[6]) - 2*d_y
            if int(dim[3]) > d_y :
                y_min = int(dim[3]) - d_y
                y_max = y_min + int(dim[6]) + 2*d_y
                if y_max > 511:
                    y_max = 511
                    y_min = 511 - int(dim[6]) - 2*d_y
            else:
                y_min = int(dim[3])
                y_max = y_min + int(dim[6]) + 2*d_y
                if y_max > 511:
                    y_max = 511
                    y_min = 511 - int(dim[6]) - 2*d_y

    return int(x_min),int(x_max),int(y_min),int(y_max)
```

# Cross Validation

```python
from glob import glob
import os
from sklearn.model_selection import train_test_split
import numpy as np
import albumentations as A
import torch
from torch.utils.data import Dataset, DataLoader
import csv
from sklearn.model_selection import GroupShuffleSplit
import pandas as pd

MEAN = [0.485, 0.456, 0.406]
STD  = [0.229, 0.224, 0.225]

def separation_fold(dataset,tsize,seed=42):
    splitter = GroupShuffleSplit(test_size=tsize, n_splits=5, random_state = seed)
    split = splitter.split(dataset, groups=dataset["patient"])
    kfold_1, kfold_2 = next(split)
    kfold1 = dataset.iloc[kfold_1]
    kfold2 = dataset.iloc[kfold_2]
    #Verification
    kfold1_patients=np.unique(kfold1["patient"])
    kfold2_patients=np.unique(kfold2["patient"])
    count=0 
    for patient in kfold1_patients:
        if patient in kfold2_patients:
            count+=1
    assert count == 0, "Bad Separations"
    return kfold1,kfold2
    
def separation(lista_total):
    train,kfold1=separation_fold(lista_total,0.2)
    kfold23,kfold45=separation_fold(train,0.5)
    kfold2,kfold3=separation_fold(kfold23,0.5)
    kfold4,kfold5=separation_fold(kfold45,0.5)
    kfold1=list(kfold1["name"])
    kfold2=list(kfold2["name"])
    kfold3=list(kfold3["name"])
    kfold4=list(kfold4["name"])
    kfold5=list(kfold5["name"])
    count=0
    for i in range(5):
        kfolds=[kfold1,kfold2,kfold3,kfold4,kfold5]
        ktest=kfolds.pop(i)
        for k in range(4):
            for j in ktest:
                if j in kfolds[k]:
                    count+=1
                    print(j)
    assert count == 0, "Bad Separations"
    #print_len(kfold1,kfold2,kfold3,kfold4,kfold5)
    return [kfold1,kfold2,kfold3,kfold4,kfold5]

def kfold_group(data,kfold,seed=42):
    data_info=[]
    for name in data:
        info={}
        info["name"]=name
        name_wp=name.split(".")[0].split("-")
        info["patient"]=name_wp[2].split("_")[0]
        info["slide"]=name_wp[3]
        data_info.append(info)
    pd_data=pd.DataFrame.from_dict(data_info)
    kfolds=separation(pd_data)
    kval=kfolds.pop(kfold-1)
    ktrain=[]
    for i in kfolds:
        ktrain.extend(i)
    np.random.seed(seed)
    np.random.shuffle(kval)
    np.random.shuffle(ktrain)
    return kval,ktrain
```

```python
class SegmentationDataset(Dataset):
    
    def __init__(self, kfold, root, transforms, preprocessing, mode = 'Train'):
        self.root = root
        
        if mode == 'Test':
            img_ids = glob(os.path.join(root, 'images', '*.npy'))
            img_ids = [(os.path.splitext(os.path.basename(p))[0]+'.npy') for p in img_ids]
            self.img_ids = img_ids
        else:
            with open(root + "data_5c.csv", "r") as archivo:
                reader = csv.reader(archivo)
                lista_total = []
                for fila in reader:
                    lista_total.append(fila[0])
            val_img_ids, train_img_ids  = kfold_group(lista_total,kfold,seed=42)
            self.img_ids = train_img_ids if mode == 'Train' else val_img_ids
        self.img_path = os.path.join(root, 'images')
        self.lbl_path = os.path.join(root, 'labels')
        self.transforms = transforms
        self.preprocessing = preprocessing

    def __getitem__(self, idx):
        # load images ad masks
        img_id = self.img_ids[idx]
        img = (np.load(os.path.join(self.img_path, img_id))).astype(np.float32) # (h,w) RGB - float32
        img = img - 1024.0
        img_min, img_max = -100, 400 #  (-17,201) # Estadistica de los datos (0.5%, 99.5%) pixeles que pertecen a las clases
        img = np.clip(img, img_min, img_max)
        # Opcion 1: Estadistica de los datos -- Entrenas en modelo sin pesos pre-entrenados
        #img_mean, img_std = 108.507, 64.996 
        #img = (img - img_mean) / img_std
        # Opcion 2: Entre 0 y 1 para aumentación de datos Y/O normalizar respecto a pesos de Imagenet
        img = (img - img_min) / (img_max - img_min) # [0,1]
        img = img[..., None] # (h,w,1)
        img = img.repeat(3, axis=-1) # (h,w,3)
        mask = np.load(os.path.join(self.lbl_path, img_id)) # (h,w) - uint8
        mask = np.expand_dims(mask, axis = -1) # (h,w,1)
        
        # apply augmentations
        if self.transforms is not None:
            augmented= self.transforms(image=img, mask=mask)
            img, mask = augmented['image'], augmented['mask'] # (h,w,c), (h,w,c)
        
        # apply preprocessing
        if self.preprocessing is not None:
            sample = self.preprocessing(image=img, mask=mask) # [0,1]
            img, mask = sample['image'], sample['mask']    # (h,w,c), (h,w,c)
        else:
            img = ((img) - MEAN ) / STD  # ImageNet
        
        # (h,w,c) -> (c,h,w)  
        img  = np.transpose(img,  (2,0,1)).astype(np.float32) #(c,h,w)
        mask = np.transpose(mask, (2,0,1)).astype(np.float32) #(c,h,w)
        
        img= torch.Tensor(img)      
        target = torch.Tensor(mask) # [0,1,2,3,4] (1-liver,2-hepatic,3-portal,4-tumor)
        return img, target

    def __len__(self):
        return len(self.img_ids)
```

```python
class SegmentationDataset(Dataset):
    
    def __init__(self, kfold, root, transforms, preprocessing, mode = 'Train'):
        self.root = root
        
        if mode == 'Test':
            img_ids = glob(os.path.join(root, 'images', '*.npy'))
            img_ids = [(os.path.splitext(os.path.basename(p))[0]+'.npy') for p in img_ids]
            self.img_ids = img_ids
        else:
            with open(root + "data_vessel_3cut_size.csv", "r") as archivo:
                reader = csv.reader(archivo)
                lista_total = []
                for fila in reader:
                    lista_total.append(fila[0])
            val_img_ids, train_img_ids  = kfold_group(lista_total,kfold,seed=42)
            self.img_ids = train_img_ids if mode == 'Train' else val_img_ids
        self.img_path = os.path.join(root, 'images')
        self.lbl_path = os.path.join(root, 'masks_vss')
        self.lbl_lv_path = os.path.join(root, 'masks_lv')
        self.transforms = transforms
        self.preprocessing = preprocessing

    def __getitem__(self, idx):
        # load images ad masks
        img_data = self.img_ids[idx].split(';')
        img_id,x_min,x_max,y_min,y_max  = img_data[0],int(img_data[1]),int(img_data[2]),int(img_data[3]),int(img_data[4])
        list_slide = f_list_3cut(img_id)
        img0 = (np.load(os.path.join(self.img_path, list_slide[0]))).astype(np.float32) - 1024.0 # (h,w) RGB - float32
        mask_lv = np.load(os.path.join(self.lbl_lv_path, list_slide[0])) # (h,w) - uint8
        img0 = img0*mask_lv
        img1 = (np.load(os.path.join(self.img_path, img_id))).astype(np.float32) - 1024.0 # (h,w) RGB - float32
        mask_lv = np.load(os.path.join(self.lbl_lv_path, img_id)) # (h,w) - uint8
        img1 = img1*mask_lv
        img2 = (np.load(os.path.join(self.img_path, list_slide[1]))).astype(np.float32) - 1024.0 # (h,w) RGB - float32
        mask_lv = np.load(os.path.join(self.lbl_lv_path, list_slide[1])) # (h,w) - uint8
        img2 = img2*mask_lv
        img0 = img0[y_min:y_max+1,x_min:x_max+1]
        img1 = img1[y_min:y_max+1,x_min:x_max+1]
        img2 = img2[y_min:y_max+1,x_min:x_max+1]
        img_min, img_max = -100, 400 #  (-17,201) # Estadistica de los datos (0.5%, 99.5%) pixeles que pertecen a las clases
        img0 = np.clip(img0, img_min, img_max)
        img1 = np.clip(img1, img_min, img_max)
        img2 = np.clip(img2, img_min, img_max)
        # Opcion 1: Estadistica de los datos -- Entrenas en modelo sin pesos pre-entrenados
        #img_mean, img_std = 108.507, 64.996 
        #img0 = (img0 - img_mean) / img_std
        #img1 = (img1 - img_mean) / img_std
        #img2 = (img2 - img_mean) / img_std
        # Opcion 2: Entre 0 y 1 para aumentación de datos Y/O normalizar respecto a pesos de Imagenet
        img0 = (img0 - img_min) / (img_max - img_min) # [0,1]
        img1 = (img1 - img_min) / (img_max - img_min) # [0,1]
        img2 = (img2 - img_min) / (img_max - img_min) # [0,1]
        img = np.dstack((img0, img1, img2)) # (h,w,3)
        mask = np.load(os.path.join(self.lbl_path, img_id)) # (h,w) - uint8
        mask = mask[y_min:y_max+1,x_min:x_max+1]
        mask[mask == 2] = 1 # vessel
        mask = np.expand_dims(mask, axis = -1) # (h,w,1)
        
        # apply augmentations
        if self.transforms is not None:
            augmented= self.transforms(image=img, mask=mask)
            img, mask = augmented['image'], augmented['mask'] # (h,w,c), (h,w,c)
        
        # apply preprocessing
        if self.preprocessing is not None:
            sample = self.preprocessing(image=img, mask=mask) # [0,1]
            img, mask = sample['image'], sample['mask']    # (h,w,c), (h,w,c)
        else:
            img = ((img) - MEAN ) / STD  # ImageNet
        
        # (h,w,c) -> (c,h,w)  
        img  = np.transpose(img,  (2,0,1)).astype(np.float32) #(c,h,w)
        mask = np.transpose(mask, (2,0,1)).astype(np.float32) #(c,h,w)
        
        img= torch.Tensor(img)      
        target = torch.Tensor(mask) # [0,1] (0-bg,1-hepatic,1-portal)
        return img, target

    def __len__(self):
        return len(self.img_ids)
```
